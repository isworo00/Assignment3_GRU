# Data Visualize

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "GOOG.csv"
data = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')

# Drop the 'Close' column
data.drop(columns=['Close'], inplace=True)

# Sort data by Date (if not already sorted)
data.sort_values(by='Date', inplace=True)

# Set Date as the index for better plotting
data.set_index('Date', inplace=True)

# Create a figure with subplots: volume spans 2 columns
fig = plt.figure(figsize=(12, 12))

# Create gridspec for custom layout
from matplotlib.gridspec import GridSpec
gs = GridSpec(4, 2, figure=fig)

# Plot price-related features in 2x2 grid
features = ['Open', 'High', 'Low', 'Adjusted Close']
titles = [
    'Open Prices Over Time',
    'High Prices Over Time',
    'Low Prices Over Time',
    'Adjusted Close Prices Over Time'
]
ylabels = ['Price', 'Price', 'Price', 'Price']

# Iterate through the first 4 features and plot them
for i, (feature, title, ylabel) in enumerate(zip(features, titles, ylabels)):
    ax = fig.add_subplot(gs[i // 2, i % 2])  # Arrange in 2x2 grid
    ax.plot(data.index, data[feature], label=feature, linewidth=2)
    ax.set_title(title, fontsize=14)
    ax.set_xlabel('Date')
    ax.set_ylabel(ylabel)
    ax.grid()
    ax.legend()
    ax.tick_params(axis='x', rotation=45)

# Plot Volume spanning across 2 columns
ax_volume = fig.add_subplot(gs[2, :])  # Span entire row for volume
ax_volume.bar(data.index, data['Volume'], color='blue', alpha=0.6, width=1.0)
ax_volume.set_title('Trading Volume Over Time', fontsize=14)
ax_volume.set_xlabel('Date')
ax_volume.set_ylabel('Volume')
ax_volume.grid(axis='y')
ax_volume.tick_params(axis='x', rotation=45)

# Adjust layout to prevent overlap
plt.tight_layout()

# Save the figure
plt.savefig('Stock_Features_With_Volume_Spanning_Two_Columns.png', format='png', dpi=300)  # High resolution for papers

# Show the figure
plt.show()
```

# Training Function

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import time
from sklearn.preprocessing import MinMaxScaler, RobustScaler
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import TimeSeriesSplit
import matplotlib.pyplot as plt
import itertools
import csv
import random
import os

# Set random seeds for reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

set_seed(42)

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')

# Set the device to use GPU if available, otherwise CPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Step 1: Load the data with date parsing
data = pd.read_csv('GOOG.csv', parse_dates=['Date'], dayfirst=True)

# Step 2: Drop the 'Close' column as we'll use 'Adjusted Close'
data = data.drop(columns=['Close'])

# Step 3: Sort data by date to ensure it's in chronological order
data = data.sort_values('Date').reset_index(drop=True)

# Step 4: Split the data into training/validation and test sets
n = len(data)
test_size = int(n * 0.125)
train_val_data = data.iloc[:-test_size].reset_index(drop=True)
test_data = data.iloc[-test_size:].reset_index(drop=True)

# Features to scale
features = ['Low', 'Open', 'Volume', 'High', 'Adjusted Close']

# Function to create sequences of data for GRU input
def create_sequences(data, seq_length, pred_length):
    xs = []
    ys = {feature: [] for feature in features}
    for i in range(len(data) - seq_length - pred_length + 1):
        x = data[i:(i+seq_length), :]  # Input sequence
        xs.append(x)
        for idx, feature in enumerate(features):
            y = data[(i+seq_length):(i+seq_length+pred_length), idx]  # Target sequence
            ys[feature].append(y)
    xs = np.array(xs)
    for feature in features:
        ys[feature] = np.array(ys[feature])
    return xs, ys

# Define the GRU model class with shared GRU and separate FC layers with Dropout
class GRUModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, dropout_rate, pred_length, feature_list):
        super(GRUModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.pred_length = pred_length
        self.input_size = input_size
        self.feature_list = feature_list

        # Shared GRU layer
        if self.num_layers < 2:
            # No dropout if GRU layer is less than 2
            self.gru = nn.GRU(input_size, hidden_size, num_layers, dropout=0.0, batch_first=True)
        else:
            self.gru = nn.GRU(input_size, hidden_size, num_layers, dropout=dropout_rate, batch_first=True)

        self.fc_dropout = nn.Dropout(dropout_rate)
        
        # Separate Dropout and Fully Connected layers for each feature
        self.fc_dropout_layers = nn.ModuleDict({
            feature: nn.Dropout(dropout_rate) for feature in feature_list
        })
        self.fc_layers = nn.ModuleDict({
            feature: nn.Linear(hidden_size, pred_length) for feature in feature_list
        })

    def forward(self, x):
        # Initialize hidden state with zeros
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(device)

        # Forward propagate GRU
        out, _ = self.gru(x, h0)  # out: (batch_size, seq_length, hidden_size)

        # Take the output from the last time step
        out = out[:, -1, :]  # Shape: (batch_size, hidden_size)

        # Apply dropout after GRU
        out = self.fc_dropout(out)

        # Output dictionary to store predictions for each feature
        outputs = {}
        for feature in self.feature_list:
            # Apply dropout before FC layer
            fc_out = self.fc_dropout_layers[feature](out)
            outputs[feature] = self.fc_layers[feature](fc_out)  # Shape: (batch_size, pred_length)

        return outputs


def training(train_val_data,
            batch_sizes,
            learning_rates,
            num_layers_list,
            hidden_sizes,
            dropout_rates,
            num_epochs_list,
            seq_lengths,
            pred_lengths,
            optimizers,
            loss_functions,
            weight_decays,
            patience_list = 10,
            n_splits = 10,
            resultslogname = 'model_results_cv.csv',
            modelname = 'model'):
    
    # Create all combinations of hyperparameters
    hyperparameter_list = list(itertools.product(
        batch_sizes,
        learning_rates,
        num_layers_list,
        hidden_sizes,
        dropout_rates,
        num_epochs_list,
        seq_lengths,
        pred_lengths,
        optimizers,
        loss_functions,
        weight_decays,
        patience_list
    ))

    # Initialize TimeSeriesSplit
    tscv = TimeSeriesSplit(n_splits=n_splits)

    # Ensure output directories exist
    os.makedirs('per_epoch_logs', exist_ok=True)
    os.makedirs('models', exist_ok=True)  # Directory to save models

    # Open a CSV file to record the aggregated results of each model
    csv_file = open(resultslogname, 'w', newline='')
    csv_writer = csv.writer(csv_file)

    # Prepare CSV header
    csv_headers = [
        'BatchSize', 'LearningRate', 'NumLayers', 'HiddenSize', 'DropoutRate',
        'NumEpochs', 'SeqLength', 'PredLength', 'Optimizer', 'LossFunction', 'WeightDecay',
        'Patience', 'NumParameters'
    ]

    # Add per-feature validation metrics to the header
    for feature in features:
        csv_headers.extend([
            f'AvgValMSE_{feature}', f'StdValMSE_{feature}',
            f'AvgValRMSE_{feature}', f'StdValRMSE_{feature}',
            f'AvgValMAE_{feature}', f'StdValMAE_{feature}',
            f'AvgValMAPE_{feature}', f'StdValMAPE_{feature}',
            f'AvgValR2_{feature}', f'StdValR2_{feature}'
        ])

    # Add per-feature training metrics to the header
    for feature in features:
        csv_headers.extend([
            f'AvgTrainMSE_{feature}', f'StdTrainMSE_{feature}',
            f'AvgTrainRMSE_{feature}', f'StdTrainRMSE_{feature}',
            f'AvgTrainMAE_{feature}', f'StdTrainMAE_{feature}',
            f'AvgTrainMAPE_{feature}', f'StdTrainMAPE_{feature}',
            f'AvgTrainR2_{feature}', f'StdTrainR2_{feature}'
        ])

    # Add overall average metrics and timings
    csv_headers.extend([
        'AvgValMSE_All', 'StdValMSE_All',
        'AvgValRMSE_All', 'StdValRMSE_All',
        'AvgValMAE_All', 'StdValMAE_All',
        'AvgValMAPE_All', 'StdValMAPE_All',
        'AvgValR2_All', 'StdValR2_All',
        'AvgTrainTimePerFold', 'StdTrainTimePerFold',
        'AvgValTimePerFold', 'StdValTimePerFold',
        'AvgTrainTimePerEpoch', 'StdTrainTimePerEpoch',
        'AvgValTimePerEpoch', 'StdValTimePerEpoch',
        'TotalTrainTime', 'TotalValTime'
    ])

    # Write the header row
    csv_writer.writerow(csv_headers)

    # Loop over all combinations of hyperparameters
    for idx, (batch_size, learning_rate, num_layers, hidden_size, dropout_rate,
            num_epochs, seq_length, pred_length, optimizer_name, loss_function_name, weight_decay, patience) in enumerate(hyperparameter_list):

        print(f"\nTraining model {idx+1}/{len(hyperparameter_list)} with hyperparameters:")
        print(f"Batch size: {batch_size}, Learning rate: {learning_rate}, Num layers: {num_layers}, "
            f"Hidden size: {hidden_size}, Dropout rate: {dropout_rate}, Num epochs: {num_epochs}, "
            f"Sequence length: {seq_length}, Prediction length: {pred_length}, "
            f"Optimizer: {optimizer_name}, Loss function: {loss_function_name}, Weight Decay: {weight_decay}, "
            f"Patience: {patience}")

        # Convert DataFrame to pandas DataFrame for fold operations
        data_array = train_val_data[features]

        # Initialize dictionaries to store metrics per feature
        train_metrics_per_feature = {feature: {'mse': [], 'rmse': [], 'mae': [], 'mape': [], 'r2': []} for feature in features}
        val_metrics_per_feature = {feature: {'mse': [], 'rmse': [], 'mae': [], 'mape': [], 'r2': []} for feature in features}

        # Initialize lists to store overall training and validation times per fold
        all_train_times_per_fold = []
        all_val_times_per_fold = []

        # Initialize lists to store training and validation times per epoch across all folds
        all_train_times_per_epoch = []
        all_val_times_per_epoch = []

        # Initialize variables to track total training and validation time across all folds
        total_train_time = 0.0
        total_val_time = 0.0

        for fold, (train_index, val_index) in enumerate(tscv.split(data_array)):
            print(f"  Fold {fold+1}/{n_splits}")

            # Get the training and validation data for this fold
            train_data_fold = data_array.iloc[train_index].reset_index(drop=True)
            val_data_fold = data_array.iloc[val_index].reset_index(drop=True)

            # Initialize scalers per feature
            scalers = {}
            for feature in features:
                if feature == 'Volume':
                    scaler = RobustScaler()
                else:
                    scaler = MinMaxScaler()
                # Fit the scaler on the training data for this feature
                scaler.fit(train_data_fold[[feature]])
                scalers[feature] = scaler

            # Transform training and validation data
            train_scaled_features = []
            val_scaled_features = []
            for feature in features:
                train_scaled_feature = scalers[feature].transform(train_data_fold[[feature]])
                val_scaled_feature = scalers[feature].transform(val_data_fold[[feature]])
                train_scaled_features.append(train_scaled_feature)
                val_scaled_features.append(val_scaled_feature)
            # Concatenate the scaled features along the columns
            train_scaled = np.hstack(train_scaled_features)
            val_scaled = np.hstack(val_scaled_features)

            # Create sequences for training and validation
            train_sequences, train_targets = create_sequences(train_scaled, seq_length, pred_length)
            val_sequences, val_targets = create_sequences(val_scaled, seq_length, pred_length)

            # If the validation set is too small to create sequences, skip this fold
            if len(train_sequences) == 0 or len(val_sequences) == 0:
                print("    Not enough data to create sequences for this fold. Skipping.")
                continue

            # Convert the sequences to PyTorch tensors
            X_train = torch.from_numpy(train_sequences).float().to(device)
            X_val = torch.from_numpy(val_sequences).float().to(device)

            # Convert targets to tensors per feature
            y_train = {feature: torch.from_numpy(train_targets[feature]).float().to(device) for feature in features}
            y_val = {feature: torch.from_numpy(val_targets[feature]).float().to(device) for feature in features}

            # Initialize the model with current hyperparameters
            input_size = len(features)
            model = GRUModel(input_size=input_size, hidden_size=hidden_size,
                            num_layers=num_layers, dropout_rate=dropout_rate,
                            pred_length=pred_length, feature_list=features)
            model = model.to(device)

            # Count the number of trainable parameters in the model
            num_parameters = sum(p.numel() for p in model.parameters() if p.requires_grad)

            # Define the loss function
            if loss_function_name == 'MSELoss':
                criterion = nn.MSELoss()
            elif loss_function_name == 'L1Loss':
                criterion = nn.L1Loss()
            elif loss_function_name == 'SmoothL1Loss':
                criterion = nn.SmoothL1Loss()
            else:
                raise ValueError("Invalid loss function")

            # Define the optimizer with selective weight decay
            decay = []
            no_decay = []
            for name, param in model.named_parameters():
                if 'bias' in name or 'norm' in name or 'LayerNorm' in name:
                    no_decay.append(param)
                else:
                    decay.append(param)

            if optimizer_name == 'Adam':
                optimizer = torch.optim.Adam([
                    {'params': decay, 'weight_decay': weight_decay},
                    {'params': no_decay, 'weight_decay': 0.0}
                ], lr=learning_rate)
            elif optimizer_name == 'SGD':
                optimizer = torch.optim.SGD([
                    {'params': decay, 'weight_decay': weight_decay},
                    {'params': no_decay, 'weight_decay': 0.0}
                ], lr=learning_rate)
            elif optimizer_name == 'AdamW':
                optimizer = torch.optim.AdamW([
                    {'params': decay, 'weight_decay': weight_decay},
                    {'params': no_decay, 'weight_decay': 0.0}
                ], lr=learning_rate)
            elif optimizer_name == 'RMSprop':
                optimizer = torch.optim.RMSprop([
                    {'params': decay, 'weight_decay': weight_decay},
                    {'params': no_decay, 'weight_decay': 0.0}
                ], lr=learning_rate)
            else:
                raise ValueError("Invalid optimizer")

            # Implement Learning Rate Scheduler
            scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)

            # Create DataLoader for the training data
            train_dataset = torch.utils.data.TensorDataset(X_train, *[y_train[feature] for feature in features])
            train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=False)

            # Early Stopping variables
            best_val_loss = float('inf')
            epochs_no_improve = 0
            early_stop = False
            best_model_state = None

            # Prepare CSV for per-epoch logging
            per_epoch_csv_filename = f'per_epoch_logs/{modelname}_{idx+1}_fold_{fold+1}_per_epoch_results.csv'
            per_epoch_csv_file = open(per_epoch_csv_filename, 'w', newline='')
            per_epoch_csv_writer = csv.writer(per_epoch_csv_file)

            # Prepare per-epoch CSV headers
            per_epoch_headers = ['Epoch', 'TrainLoss', 'ValLoss']
            for feature in features:
                per_epoch_headers.extend([
                    f'TrainMSE_{feature}', f'ValMSE_{feature}',
                    f'TrainRMSE_{feature}', f'ValRMSE_{feature}',
                    f'TrainMAE_{feature}', f'ValMAE_{feature}',
                    f'TrainMAPE_{feature}', f'ValMAPE_{feature}',
                    f'TrainR2_{feature}', f'ValR2_{feature}'
                ])
            # Add training and validation time per epoch
            per_epoch_headers.extend(['TrainTime', 'ValTime'])

            # Write headers
            per_epoch_csv_writer.writerow(per_epoch_headers)

            # Initialize lists to store per-epoch times
            epoch_train_times = []
            epoch_val_times = []

            # Initialize variables to track overall fold timing
            fold_train_time = 0.0
            fold_val_time = 0.0

            # Start training the model
            model.train()
            for epoch in range(num_epochs):
                # Start timer for training
                start_epoch_train = time.time()

                model.train()
                epoch_train_loss = 0
                for batch in train_loader:
                    batch_x = batch[0]
                    batch_y = {feature: batch[idx+1] for idx, feature in enumerate(features)}

                    # Forward pass: compute predicted y by passing x to the model
                    outputs = model(batch_x)

                    # Compute individual losses
                    feature_losses = {}
                    total_loss = 0.0
                    for feature in features:
                        loss = criterion(outputs[feature], batch_y[feature])
                        feature_losses[feature] = loss.item()
                        total_loss += loss

                    # Average loss for backpropagation
                    average_loss = total_loss / len(features)

                    # Backward pass and optimization
                    optimizer.zero_grad()
                    average_loss.backward()
                    optimizer.step()

                    # Accumulate training loss
                    epoch_train_loss += average_loss.item()

                # End timer for training
                train_time = time.time() - start_epoch_train
                epoch_train_times.append(train_time)
                fold_train_time += train_time

                # Average training loss for the epoch
                epoch_train_loss /= len(train_loader)

                # Start timer for validation
                start_epoch_val = time.time()

                # Validation step
                model.eval()
                with torch.no_grad():
                    val_outputs = model(X_val)
                    val_loss = 0.0
                    for feature in features:
                        loss = criterion(val_outputs[feature], y_val[feature])
                        val_loss += loss.item()
                    val_loss /= len(features)

                # End timer for validation
                val_time = time.time() - start_epoch_val
                epoch_val_times.append(val_time)
                fold_val_time += val_time

                # Early Stopping Check
                if val_loss < best_val_loss:
                    best_val_loss = val_loss
                    epochs_no_improve = 0
                    best_model_state = model.state_dict()  # Save the best model state
                else:
                    epochs_no_improve += 1
                    if epochs_no_improve >= patience:
                        print(f"    Early stopping at epoch {epoch+1}")
                        early_stop = True
                        break

                # Step the scheduler
                scheduler.step(val_loss)

                # Compute metrics per feature
                with torch.no_grad():
                    train_outputs = model(X_train)
                    y_train_np = {feature: y_train[feature].cpu().numpy() for feature in features}
                    train_pred_np = {feature: train_outputs[feature].cpu().numpy() for feature in features}
                    y_val_np = {feature: y_val[feature].cpu().numpy() for feature in features}
                    val_pred_np = {feature: val_outputs[feature].cpu().numpy() for feature in features}

                    # Initialize epsilon to prevent division by zero in MAPE calculation
                    epsilon = 1e-10

                    # Store metrics per feature
                    epoch_metrics = []
                    for feature in features:
                        # Training Metrics
                        y_train_flat = y_train_np[feature].reshape(-1)
                        train_pred_flat = train_pred_np[feature].reshape(-1)
                        train_mse = mean_squared_error(y_train_flat, train_pred_flat)
                        train_rmse = np.sqrt(train_mse)
                        train_mae = mean_absolute_error(y_train_flat, train_pred_flat)
                        train_mape = np.mean(np.abs((y_train_flat - train_pred_flat) / (y_train_flat + epsilon))) * 100
                        train_r2 = r2_score(y_train_flat, train_pred_flat)

                        # Validation Metrics
                        y_val_flat = y_val_np[feature].reshape(-1)
                        val_pred_flat = val_pred_np[feature].reshape(-1)
                        val_mse = mean_squared_error(y_val_flat, val_pred_flat)
                        val_rmse = np.sqrt(val_mse)
                        val_mae = mean_absolute_error(y_val_flat, val_pred_flat)
                        val_mape = np.mean(np.abs((y_val_flat - val_pred_flat) / (y_val_flat + epsilon))) * 100
                        val_r2 = r2_score(y_val_flat, val_pred_flat)

                        # Append metrics
                        epoch_metrics.extend([
                            train_mse, val_mse,
                            train_rmse, val_rmse,
                            train_mae, val_mae,
                            train_mape, val_mape,
                            train_r2, val_r2
                        ])

                        # Append to fold's loss lists
                        train_metrics_per_feature[feature]['mse'].append(train_mse)
                        train_metrics_per_feature[feature]['rmse'].append(train_rmse)
                        train_metrics_per_feature[feature]['mae'].append(train_mae)
                        train_metrics_per_feature[feature]['mape'].append(train_mape)
                        train_metrics_per_feature[feature]['r2'].append(train_r2)

                        val_metrics_per_feature[feature]['mse'].append(val_mse)
                        val_metrics_per_feature[feature]['rmse'].append(val_rmse)
                        val_metrics_per_feature[feature]['mae'].append(val_mae)
                        val_metrics_per_feature[feature]['mape'].append(val_mape)
                        val_metrics_per_feature[feature]['r2'].append(val_r2)

                # Write per-epoch metrics and times to CSV
                per_epoch_row = [epoch+1, epoch_train_loss, val_loss] + epoch_metrics + [train_time, val_time]
                per_epoch_csv_writer.writerow(per_epoch_row)
                per_epoch_csv_file.flush()

                '''
                print(f"    Epoch {epoch+1}, Train Loss: {epoch_train_loss:.4f}, Val Loss: {val_loss:.4f}")
                for feature in features:
                    print(f"        Feature: {feature}")
                    print(f"            Train MSE: {train_mse:.4f}, Val MSE: {val_mse:.4f}")
                    print(f"            Train RMSE: {train_rmse:.4f}, Val RMSE: {val_rmse:.4f}")
                    print(f"            Train MAE: {train_mae:.4f}, Val MAE: {val_mae:.4f}")
                    print(f"            Train MAPE: {train_mape:.2f}%, Val MAPE: {val_mape:.2f}%")
                    print(f"            Train R²: {train_r2:.4f}, Val R²: {val_r2:.4f}")
                    print()
                print(f"    Train Time: {train_time:.2f}s, Val Time: {val_time:.2f}s")
                '''

            # Append overall fold times to the global lists
            all_train_times_per_fold.append(fold_train_time)
            all_val_times_per_fold.append(fold_val_time)

            # Append all per-epoch times from this fold to the global lists
            all_train_times_per_epoch.extend(epoch_train_times)
            all_val_times_per_epoch.extend(epoch_val_times)

            # Accumulate total training and validation times across all folds
            total_train_time += fold_train_time
            total_val_time += fold_val_time

            # Load best model state
            if best_model_state is not None:
                model.load_state_dict(best_model_state)

            # Save the best model for this fold
            model_save_path = f'models/gru_{modelname}_hyperparam_{idx+1}_fold_{fold+1}_best.pth'
            torch.save(best_model_state, model_save_path)

            # Evaluate the model on the training set with best model
            model.eval()
            with torch.no_grad():
                train_outputs = model(X_train)
                y_train_np = {feature: y_train[feature].cpu().numpy() for feature in features}
                train_pred_np = {feature: train_outputs[feature].cpu().numpy() for feature in features}

                # Initialize epsilon to prevent division by zero in MAPE calculation
                epsilon = 1e-10

                # Loop over features to compute training metrics
                for feature in features:
                    # Flatten predictions and targets
                    y_train_flat = y_train_np[feature].reshape(-1)
                    train_pred_flat = train_pred_np[feature].reshape(-1)

                    # Training metrics
                    train_mse = mean_squared_error(y_train_flat, train_pred_flat)
                    train_rmse = np.sqrt(train_mse)
                    train_mae = mean_absolute_error(y_train_flat, train_pred_flat)
                    train_mape = np.mean(np.abs((y_train_flat - train_pred_flat) / (y_train_flat + epsilon))) * 100
                    train_r2 = r2_score(y_train_flat, train_pred_flat)

                    # Append metrics
                    train_metrics_per_feature[feature]['mse'].append(train_mse)
                    train_metrics_per_feature[feature]['rmse'].append(train_rmse)
                    train_metrics_per_feature[feature]['mae'].append(train_mae)
                    train_metrics_per_feature[feature]['mape'].append(train_mape)
                    train_metrics_per_feature[feature]['r2'].append(train_r2)

            # Evaluate the model on the validation set
            with torch.no_grad():
                val_outputs = model(X_val)
                y_val_np = {feature: y_val[feature].cpu().numpy() for feature in features}
                val_pred_np = {feature: val_outputs[feature].cpu().numpy() for feature in features}

                # Loop over features to compute validation metrics
                for feature in features:
                    # Flatten predictions and targets
                    y_val_flat = y_val_np[feature].reshape(-1)
                    val_pred_flat = val_pred_np[feature].reshape(-1)

                    # Validation metrics
                    val_mse = mean_squared_error(y_val_flat, val_pred_flat)
                    val_rmse = np.sqrt(val_mse)
                    val_mae = mean_absolute_error(y_val_flat, val_pred_flat)
                    val_mape = np.mean(np.abs((y_val_flat - val_pred_flat) / (y_val_flat + epsilon))) * 100
                    val_r2 = r2_score(y_val_flat, val_pred_flat)

                    # Append metrics
                    val_metrics_per_feature[feature]['mse'].append(val_mse)
                    val_metrics_per_feature[feature]['rmse'].append(val_rmse)
                    val_metrics_per_feature[feature]['mae'].append(val_mae)
                    val_metrics_per_feature[feature]['mape'].append(val_mape)
                    val_metrics_per_feature[feature]['r2'].append(val_r2)

            per_epoch_csv_file.close()  # Close per-epoch CSV file

            print(f"  Fold {fold+1} completed.")

        # Check if any folds were processed
        if len(all_train_times_per_fold) == 0:
            print("No folds were processed. Skipping this hyperparameter combination.")
            continue

        # Aggregate metrics per feature
        avg_train_metrics_per_feature = {feature: {} for feature in features}
        std_train_metrics_per_feature = {feature: {} for feature in features}
        avg_val_metrics_per_feature = {feature: {} for feature in features}
        std_val_metrics_per_feature = {feature: {} for feature in features}

        for feature in features:
            for metric in ['mse', 'rmse', 'mae', 'mape', 'r2']:
                avg_train_metrics_per_feature[feature][metric] = np.mean(train_metrics_per_feature[feature][metric])
                std_train_metrics_per_feature[feature][metric] = np.std(train_metrics_per_feature[feature][metric])
                avg_val_metrics_per_feature[feature][metric] = np.mean(val_metrics_per_feature[feature][metric])
                std_val_metrics_per_feature[feature][metric] = np.std(val_metrics_per_feature[feature][metric])

        # Calculate overall average metrics across all features
        avg_val_mse_all = np.mean([avg_val_metrics_per_feature[feature]['mse'] for feature in features])
        std_val_mse_all = np.mean([std_val_metrics_per_feature[feature]['mse'] for feature in features])
        avg_val_rmse_all = np.mean([avg_val_metrics_per_feature[feature]['rmse'] for feature in features])
        std_val_rmse_all = np.mean([std_val_metrics_per_feature[feature]['rmse'] for feature in features])
        avg_val_mae_all = np.mean([avg_val_metrics_per_feature[feature]['mae'] for feature in features])
        std_val_mae_all = np.mean([std_val_metrics_per_feature[feature]['mae'] for feature in features])
        avg_val_mape_all = np.mean([avg_val_metrics_per_feature[feature]['mape'] for feature in features])
        std_val_mape_all = np.mean([std_val_metrics_per_feature[feature]['mape'] for feature in features])
        avg_val_r2_all = np.mean([avg_val_metrics_per_feature[feature]['r2'] for feature in features])
        std_val_r2_all = np.mean([std_val_metrics_per_feature[feature]['r2'] for feature in features])

        # Calculate average and std of overall training and validation times per fold
        avg_train_time_per_fold = np.mean(all_train_times_per_fold)
        std_train_time_per_fold = np.std(all_train_times_per_fold)
        avg_val_time_per_fold = np.mean(all_val_times_per_fold)
        std_val_time_per_fold = np.std(all_val_times_per_fold)

        # Calculate average and std of training and validation times per epoch
        avg_train_time_per_epoch = np.mean(all_train_times_per_epoch)
        std_train_time_per_epoch = np.std(all_train_times_per_epoch)
        avg_val_time_per_epoch = np.mean(all_val_times_per_epoch)
        std_val_time_per_epoch = np.std(all_val_times_per_epoch)

        # Calculate total training and validation time across all folds
        total_train_time = sum(all_train_times_per_fold)
        total_val_time = sum(all_val_times_per_fold)

        # Prepare data row for CSV
        row_data = [
            batch_size, learning_rate, num_layers, hidden_size, dropout_rate,
            num_epochs, seq_length, pred_length, optimizer_name, loss_function_name, weight_decay,
            patience, num_parameters
        ]

        # Add per-feature validation metrics to the row
        for feature in features:
            row_data.extend([
                avg_val_metrics_per_feature[feature]['mse'], std_val_metrics_per_feature[feature]['mse'],
                avg_val_metrics_per_feature[feature]['rmse'], std_val_metrics_per_feature[feature]['rmse'],
                avg_val_metrics_per_feature[feature]['mae'], std_val_metrics_per_feature[feature]['mae'],
                avg_val_metrics_per_feature[feature]['mape'], std_val_metrics_per_feature[feature]['mape'],
                avg_val_metrics_per_feature[feature]['r2'], std_val_metrics_per_feature[feature]['r2']
            ])

        # Add per-feature training metrics to the row
        for feature in features:
            row_data.extend([
                avg_train_metrics_per_feature[feature]['mse'], std_train_metrics_per_feature[feature]['mse'],
                avg_train_metrics_per_feature[feature]['rmse'], std_train_metrics_per_feature[feature]['rmse'],
                avg_train_metrics_per_feature[feature]['mae'], std_train_metrics_per_feature[feature]['mae'],
                avg_train_metrics_per_feature[feature]['mape'], std_train_metrics_per_feature[feature]['mape'],
                avg_train_metrics_per_feature[feature]['r2'], std_train_metrics_per_feature[feature]['r2']
            ])

        # Add overall average metrics and timings
        row_data.extend([
            avg_val_mse_all, std_val_mse_all,
            avg_val_rmse_all, std_val_rmse_all,
            avg_val_mae_all, std_val_mae_all,
            avg_val_mape_all, std_val_mape_all,
            avg_val_r2_all, std_val_r2_all,
            avg_train_time_per_fold, std_train_time_per_fold,
            avg_val_time_per_fold, std_val_time_per_fold,
            avg_train_time_per_epoch, std_train_time_per_epoch,
            avg_val_time_per_epoch, std_val_time_per_epoch,
            total_train_time, total_val_time
        ])

        # Write the results and hyperparameters to the CSV file
        csv_writer.writerow(row_data)

        # Flush the CSV file to ensure data is written
        csv_file.flush()

        '''
        for feature in features:
            print(f"  Metrics for Feature: {feature}")
            print(f"    Avg Val MSE: {avg_val_metrics_per_feature[feature]['mse']:.4f}, Std Val MSE: {std_val_metrics_per_feature[feature]['mse']:.4f}")
            print(f"    Avg Val RMSE: {avg_val_metrics_per_feature[feature]['rmse']:.4f}, Std Val RMSE: {std_val_metrics_per_feature[feature]['rmse']:.4f}")
            print(f"    Avg Val MAE: {avg_val_metrics_per_feature[feature]['mae']:.4f}, Std Val MAE: {std_val_metrics_per_feature[feature]['mae']:.4f}")
            print(f"    Avg Val MAPE: {avg_val_metrics_per_feature[feature]['mape']:.2f}%, Std Val MAPE: {std_val_metrics_per_feature[feature]['mape']:.2f}%")
            print(f"    Avg Val R2: {avg_val_metrics_per_feature[feature]['r2']:.4f}, Std Val R2: {std_val_metrics_per_feature[feature]['r2']:.4f}")
            print()

        print(f"  Average Train Time per Fold: {avg_train_time_per_fold:.2f}s, Std Train Time per Fold: {std_train_time_per_fold:.2f}s")
        print(f"  Average Val Time per Fold: {avg_val_time_per_fold:.2f}s, Std Val Time per Fold: {std_val_time_per_fold:.2f}s")
        print(f"  Average Train Time per Epoch: {avg_train_time_per_epoch:.2f}s, Std Train Time per Epoch: {std_train_time_per_epoch:.2f}s")
        print(f"  Average Val Time per Epoch: {avg_val_time_per_epoch:.2f}s, Std Val Time per Epoch: {std_val_time_per_epoch:.2f}s")
        print(f"  Total Train Time: {total_train_time:.2f}s, Total Val Time: {total_val_time:.2f}s")
        '''

    # Close the CSV file after all models have been trained
    csv_file.close()




import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

def plot_validation_loss_grouped(csv_path='model_results_cv.csv', output_dir='validation_loss_plots', hyperparameter='BatchSize'):
    """
    This function reads the model_results_cv.csv file and plots the average validation loss
    for all features based on a specified hyperparameter. It creates a grouped bar chart
    where each feature has bars representing different hyperparameter values, colored based
    on their corresponding loss values.

    Parameters:
    - csv_path (str): Path to the CSV file containing model results.
    - output_dir (str): Directory where the plots will be saved.
    - hyperparameter (str): The hyperparameter to use for comparisons (e.g., 'BatchSize').
    """
    import os
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns

    # Check if the CSV file exists
    if not os.path.exists(csv_path):
        print(f"Error: The file '{csv_path}' does not exist. Please ensure the file path is correct.")
        return

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Load the aggregated results
    results_df = pd.read_csv(csv_path)

    # Define features
    features = ['Low', 'Open', 'Volume', 'High', 'Adjusted Close']
    price_features = ['Low', 'Open', 'High', 'Adjusted Close']

    # Add new metrics
    results_df['AvgValMSE_Prices Average'] = results_df[[f'AvgValMSE_{feature}' for feature in price_features]].mean(axis=1)
    results_df['StdValMSE_Prices Average'] = results_df[[f'StdValMSE_{feature}' for feature in price_features]].mean(axis=1)
    results_df['AvgValMSE_All Features'] = results_df[[f'AvgValMSE_{feature}' for feature in features]].mean(axis=1)
    results_df['StdValMSE_All Features'] = results_df[[f'StdValMSE_{feature}' for feature in features]].mean(axis=1)

    # Extend features
    extended_features = ['Low', 'Open', 'Volume', 'High', 'Adjusted Close', 'Prices Average', 'All Features']

    # Extract relevant columns for plotting
    plot_data = results_df[[hyperparameter] + [f'AvgValMSE_{feature}' for feature in extended_features]].copy()
    plot_data.sort_values(hyperparameter, inplace=True)

    # Melt the dataframe to long format for seaborn
    melted_df = plot_data.melt(id_vars=[hyperparameter], 
                               value_vars=[f'AvgValMSE_{feature}' for feature in extended_features],
                               var_name='Feature',
                               value_name='AvgValMSE')

    # Clean the 'Feature' column to remove the prefix
    melted_df['Feature'] = melted_df['Feature'].str.replace('AvgValMSE_', '')

    # Set up the matplotlib figure
    plt.figure(figsize=(14, 8))
    sns.set(style="whitegrid")

    # Choose a highly contrasting color palette
    # Options: 'tab10', 'tab20', 'Set2', or a custom palette
    palette = sns.color_palette("tab10")  # You can replace "tab10" with other palettes

    # Create a grouped barplot
    ax = sns.barplot(data=melted_df, x='Feature', y='AvgValMSE', hue=hyperparameter, palette=palette)

    # Set plot labels and title
    ax.set_title(f'Average Validation Loss by Feature and {hyperparameter}', fontsize=18)
    ax.set_xlabel('Feature', fontsize=14)
    ax.set_ylabel('Average Validation MSE', fontsize=14)

    # Rotate x-axis labels for readability
    plt.xticks()

    # Adjust legend title and position
    plt.legend(title=hyperparameter, bbox_to_anchor=(1.05, 1), loc='upper left')

    # Tight layout for better spacing
    plt.tight_layout()

    # Save the plot
    combined_plot_path = os.path.join(output_dir, f'Grouped_AvgValMSE_{hyperparameter}.png')
    plt.savefig(combined_plot_path, dpi=300)
    print(f"Grouped bar plot saved to {combined_plot_path}")
    plt.show()
```

```python
def plot_validation_loss(csv_path='model_results_cv.csv', output_dir='validation_loss_plots', hyperparameter='BatchSize'):
    """
    This function reads the model_results_cv.csv file and plots the average validation loss
    for all models based on a specified hyperparameter, using subplots for better readability.

    Parameters:
    - csv_path (str): Path to the CSV file containing model results.
    - output_dir (str): Directory where the plots will be saved.
    - hyperparameter (str): The hyperparameter to use for comparisons (e.g., 'BatchSize').
    """
    import os
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np

    # Check if the CSV file exists
    if not os.path.exists(csv_path):
        print(f"Error: The file '{csv_path}' does not exist. Please ensure the file path is correct.")
        return

    # Create the output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Load the aggregated results
    results_df = pd.read_csv(csv_path)

    # Define features
    features = ['Low', 'Open', 'Volume', 'High', 'Adjusted Close']
    price_features = ['Low', 'Open', 'High', 'Adjusted Close']

    # Add new metrics
    results_df['AvgValMSE_Prices Average'] = results_df[[f'AvgValMSE_{feature}' for feature in price_features]].mean(axis=1)
    results_df['StdValMSE_Prices Average'] = results_df[[f'StdValMSE_{feature}' for feature in price_features]].mean(axis=1)
    results_df['AvgValMSE_All Features'] = results_df[[f'AvgValMSE_{feature}' for feature in features]].mean(axis=1)
    results_df['StdValMSE_All Features'] = results_df[[f'StdValMSE_{feature}' for feature in features]].mean(axis=1)

    # Extend features
    features.extend(['Prices Average', 'All Features'])

    # Extract data for plotting
    plot_data = results_df[[hyperparameter] + [f'AvgValMSE_{feature}' for feature in features] +
                           [f'StdValMSE_{feature}' for feature in features]].copy()
    plot_data.sort_values(hyperparameter, inplace=True)

    # Set up plot parameters
    sns.set(style="whitegrid")
    num_features = len(features)

    # Create subplots
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(20, 10), sharex=True, sharey=True)
    axes = axes.flatten()

    for idx, feature in enumerate(features):
        avg_col = f'AvgValMSE_{feature}'
        std_col = f'StdValMSE_{feature}'
        positions = np.arange(len(plot_data[hyperparameter]))  # x-axis positions for the bars

        axes[idx].bar(positions, plot_data[avg_col], yerr=plot_data[std_col], capsize=5, alpha=0.8,
                      color=sns.color_palette("husl", num_features)[idx], edgecolor='black')
        axes[idx].set_title(feature, fontsize=12)
        axes[idx].set_xlabel(hyperparameter, fontsize=10)
        axes[idx].set_ylabel('AvgValMSE', fontsize=10)
        axes[idx].set_xticks(positions)
        axes[idx].set_xticklabels(plot_data[hyperparameter], fontsize=8, rotation=45)

    # Remove unused subplots if there are fewer features
    for i in range(len(features), len(axes)):
        fig.delaxes(axes[i])

    # Adjust layout
    plt.tight_layout()
    combined_plot_path = os.path.join(output_dir, f'Improved_Subplot_AvgValMSE_{hyperparameter}.png')
    plt.savefig(combined_plot_path)
    print(f"Saved to {combined_plot_path}")
    plt.show()
```

## Comparing The R2 (Base, Seq30, Seq60)

```python
import pandas as pd
import glob
import os

def summarize_r2_metrics(directory):
    baseline_pattern = "baseline_1_fold_*_per_epoch_results.csv"
    seq30_pattern = "LossFunction30_2_fold_*_per_epoch_results.csv"
    seq60_pattern = "LossFunction_3_fold_*_per_epoch_results.csv"
    
    baseline_files = glob.glob(os.path.join(directory, baseline_pattern))
    seq30_files = glob.glob(os.path.join(directory, seq30_pattern))
    seq60_files = glob.glob(os.path.join(directory, seq60_pattern))

    #print("Baseline files found:", baseline_files)
    #print("Seq30 files found:", seq30_files)
    #print("Seq60 files found:", seq60_files)

    r2_columns = [
        "ValR2_Low",
        "ValR2_Open",
        "ValR2_Volume",
        "ValR2_High",
        "ValR2_Adjusted Close"
    ]

    def aggregate_r2(files):
        r2_data = {col: [] for col in r2_columns}
        for f in files:
            df = pd.read_csv(f)
            # Only consider the last row (latest epoch)
            if len(df) > 0:
                last_row = df.iloc[-1]
                for col in r2_columns:
                    if col in df.columns:
                        val = last_row[col]
                        if pd.notna(val):
                            r2_data[col].append(val)
                    else:
                        print(f"Column {col} not found in {f}")
        return r2_data

    baseline_data = aggregate_r2(baseline_files)
    seq30_data = aggregate_r2(seq30_files)
    seq60_data = aggregate_r2(seq60_files)

    models = {
        'Base Model': baseline_data,
        'Seq30': seq30_data,
        'Seq60': seq60_data
    }

    result_data = []
    for model_name, data_dict in models.items():
        for col in r2_columns:
            values = data_dict[col]
            if len(values) == 0:
                row = [model_name, col, float('nan'), float('nan'), float('nan'), float('nan')]
            else:
                row = [
                    model_name,
                    col,
                    min(values),
                    max(values),
                    sum(values)/len(values),
                    pd.Series(values).std()
                ]
            result_data.append(row)

    summary_df = pd.DataFrame(result_data, columns=['Model', 'R2_Metric', 'Min', 'Max', 'Average', 'Std'])
    return summary_df

summary = summarize_r2_metrics('per_epoch_logs')
summary
```

```python
def evaluate():
    # Read the results from the CSV file into a DataFrame
    results_df = pd.read_csv('LossFunction_comparison_results_cv.csv')

    # Now, select the best hyperparameter combination based on a specific metric
    # For example, selecting the model with the lowest AvgValMSE for 'Adjusted Close'

    # Check if 'AvgValMSE_Adjusted Close' exists in the DataFrame
    best_metric = 'AvgValMSE_All'
    if best_metric not in results_df.columns:
        print(f"Metric {best_metric} not found in the results. Please verify the CSV headers.")
        exit()

    best_model_row = results_df.loc[results_df[best_metric].idxmin()]

    # Extract the best hyperparameters
    best_batch_size = int(best_model_row['BatchSize'])
    best_learning_rate = best_model_row['LearningRate']
    best_num_layers = int(best_model_row['NumLayers'])
    best_hidden_size = int(best_model_row['HiddenSize'])
    best_dropout_rate = best_model_row['DropoutRate']
    best_num_epochs = int(best_model_row['NumEpochs'])
    best_seq_length = int(best_model_row['SeqLength'])
    best_pred_length = int(best_model_row['PredLength'])
    best_optimizer_name = best_model_row['Optimizer']
    best_loss_function_name = best_model_row['LossFunction']
    best_weight_decay = best_model_row['WeightDecay']
    best_patience = int(best_model_row['Patience'])

    print(f"\nBest hyperparameters found based on {best_metric}:\n{best_model_row}")

    # ### Step 5: Load the Best Model from Cross-Validation

    # Assuming the best model is from the last fold of the best hyperparameter combination
    # Modify this if you have a different selection criterion

    # Extract the index of the best hyperparameter combination
    best_hyperparam_idx = results_df[best_metric].idxmin()

    # Construct the model path
    # Assuming the last fold (n_splits) was saved for the best hyperparameter combination
    best_model_path = f'models/gru_LossFunction_hyperparam_{best_hyperparam_idx+1}_fold_{5}_best.pth'

    if not os.path.exists(best_model_path):
        print(f"Best model file {best_model_path} does not exist. Please check the model saving process.")
        exit()

    # Initialize the best model
    best_model = GRUModel(input_size=len(features), 
                        hidden_size=int(best_model_row['HiddenSize']),
                        num_layers=int(best_model_row['NumLayers']),
                        dropout_rate=float(best_model_row['DropoutRate']),
                        pred_length=int(best_model_row['PredLength']),
                        feature_list=features)
    best_model.load_state_dict(torch.load(best_model_path))
    best_model = best_model.to(device)
    best_model.eval()

    # ### Step 6: Prepare the Test Set

    # Fit scalers on the entire training/validation data
    scalers = {}
    for feature in features:
        if feature == 'Volume':
            scaler = RobustScaler()
        else:
            scaler = MinMaxScaler()
        scaler.fit(train_val_data[[feature]])
        scalers[feature] = scaler

    # Transform the test data
    test_scaled_features = []
    for feature in features:
        test_scaled_feature = scalers[feature].transform(test_data[[feature]])
        test_scaled_features.append(test_scaled_feature)
    test_scaled = np.hstack(test_scaled_features)

    # Create sequences for the test set
    test_sequences, test_targets = create_sequences(test_scaled, best_seq_length, best_pred_length)

    # Convert to tensors
    X_test = torch.from_numpy(test_sequences).float().to(device)

    # Convert targets to tensors per feature
    y_test = {feature: torch.from_numpy(test_targets[feature]).float().to(device) for feature in features}

    # ### Step 7: Make Predictions on the Test Set
    with torch.no_grad():
        test_outputs = best_model(X_test)
        y_test_np = {feature: y_test[feature].cpu().numpy() for feature in features}
        test_pred_np = {feature: test_outputs[feature].cpu().numpy() for feature in features}

    # Initialize epsilon to prevent division by zero in MAPE calculation
    epsilon = 1e-10

    # Initialize dictionary to store test metrics per feature
    test_metrics_per_feature = {}

    for feature in features:
        y_test_flat = y_test_np[feature].reshape(-1)
        test_pred_flat = test_pred_np[feature].reshape(-1)

        # Inverse transform per feature
        y_test_inv = scalers[feature].inverse_transform(y_test_flat.reshape(-1,1)).reshape(-1)
        test_pred_inv = scalers[feature].inverse_transform(test_pred_flat.reshape(-1,1)).reshape(-1)

        # Compute metrics
        test_mse = mean_squared_error(y_test_inv, test_pred_inv)
        test_rmse = np.sqrt(test_mse)
        test_mae = mean_absolute_error(y_test_inv, test_pred_inv)
        test_mape = np.mean(np.abs((y_test_inv - test_pred_inv) / (y_test_inv + epsilon))) * 100
        test_r2 = r2_score(y_test_inv, test_pred_inv)

        # Store metrics
        test_metrics_per_feature[feature] = {
            'MSE': test_mse,
            'RMSE': test_rmse,
            'MAE': test_mae,
            'MAPE': test_mape,
            'R2': test_r2,
            'Actual': y_test_inv,
            'Predicted': test_pred_inv
        }

    # ### Step 8: Print Test Metrics

    for feature, metrics in test_metrics_per_feature.items():
        print(f"\nTest Metrics for {feature}:")
        print(f"  MSE: {metrics['MSE']:.4f}")
        print(f"  RMSE: {metrics['RMSE']:.4f}")
        print(f"  MAE: {metrics['MAE']:.4f}")
        print(f"  MAPE: {metrics['MAPE']:.2f}%")
        print(f"  R²: {metrics['R2']:.4f}")

    # ### Step 9: Visualization

    # Plot the predictions against the actual values on the test set for each feature
    for feature in features:
        plt.figure(figsize=(12,6))
        plt.plot(test_metrics_per_feature[feature]['Actual'], label='Actual')
        plt.plot(test_metrics_per_feature[feature]['Predicted'], label='Predicted')
        plt.title(f'Test Set - Actual vs Predicted for {feature}')
        plt.xlabel('Time')
        plt.ylabel(feature)
        plt.legend()
        plt.grid(True)
        plt.show()
```
